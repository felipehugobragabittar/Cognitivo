# 4) CASE COGNITIVO

No código abaixo foi feito:

*   Criação do dataset de treino e de teste com a regra 90-10% , onde 90% dos dados são usados para treino e 10% para teste de desempenho
*   Normalização das entradas/saidas para que não haja saturação da FUNÇÃO DE ATIVAÇÃO (tangente hiperbólica) DA RNA 


```python
y= df['Feature7']
x=df.loc[:, ~df.columns.isin(['Feature7'])]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)


norm = MinMaxScaler().fit(x_train)
x_train = norm.transform(x_train)
x_test = norm.transform(x_test)


print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```



# 5) CASE COGNITIVO

No código abaixo foi feito:

*  Testes de TOPOLOGIA das RNAS
*   foram testadas as topologias com 10, 15 e 20 neuronios na primeira camada escondida (mantendo-se a taxa de aprendizado/otimizações).



```python
clf1 = MLPRegressor(solver='adam', alpha=1e-5, activation='tanh', max_iter=1000, hidden_layer_sizes=(10,))
clf1.fit(x_train, y_train)
expected_y  = y_test
predicted_y = clf1.predict(x_test)
print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_squared_log_error(expected_y, predicted_y))



clf2 = MLPRegressor(solver='adam', alpha=1e-5, activation='tanh', max_iter=1000, hidden_layer_sizes=(15,))
clf2.fit(x_train, y_train)
expected_y  = y_test
predicted_y = clf2.predict(x_test)
print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_squared_log_error(expected_y, predicted_y))



clf3 = MLPRegressor(solver='adam', alpha=1e-5, activation='tanh', max_iter=1000, hidden_layer_sizes=(20,))
clf3.fit(x_train, y_train)
expected_y  = y_test
predicted_y = clf3.predict(x_test)
print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_squared_log_error(expected_y, predicted_y))
```
